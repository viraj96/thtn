```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time_mean_data = np.array([[0.171121,0.560946, 2.181365, 3.697824, 11.145212],
             [0.435448, 0.893944, 3.351179, 6.175125, 6.523109],
             [0.505422, 1.750609, 4.05199, 8.406093, 13.917619],
             [0.759588, 3.070201, 5.255025, 19.157942, 20.226724],
             [0.848824, 2.217931, 7.066747, 13.412251, 16.403316]])
```

```python
time_std_data = np.array([[0.026905, 0.02252, 0.079433, 0.147522, 0.344809],
                 [0.087707, 0.04924, 0.076455, 0.058406, 0.099782],
                 [0.050482, 0.077404, 0.125695, 0.142814, 0.198493],
                 [0.100972, 0.104378, 0.330358, 1.254378, 1.041322],
                 [0.127772, 0.156822, 0.48449, 0.435586, 0.732997]])
```

```python
makespan_mean_data = np.array([[538, 1070, 1540, 2010, 2548],
                      [926, 1332, 2212, 2342, 3054],
                      [1141.166667, 2034, 2408, 3640, 5192],
                      [1233.666667, 2560, 3338, 4644, 5792],
                      [1300.166667, 1980, 3790, 4982, 5300]])
```

```python
makespan_std_data = np.array([[54, 25.690465, 28.284271, 20.493902, 25.612497],
                     [113.800996, 65.238026, 78.076885, 61.611687, 49.030603], 
                     [166.779212, 56.603887, 77.562878, 119.666202, 85.883642], 
                     [213.463242, 205.523721, 110.072703, 157.936696, 114.262855], 
                     [179.582773, 210.333069, 175.556259, 239.741527, 81.975606]])
```

```python
for request in range(5):
    r_str = "Requests = " + str((request + 1) * 5)
    plt.plot(np.arange(6)[1:] * 5, time_mean_data[request, :], label = r_str, marker = "o")
    plt.fill_between(np.arange(6)[1:] * 5, time_mean_data[request, :] - time_std_data[request, :], time_mean_data[request, :] + time_std_data[request, :], alpha = 0.5)
    plt.xlabel("Number of blocks")
    plt.ylabel("Time(s)")
ax = plt.gca()
for i, label in enumerate(ax.get_xticklabels()):
    if i % 2 == 0:
        label.set_visible(False)
plt.title("Computation Time(s) Benchmarks")
plt.legend()
```

```python
for request in range(5):
    r_str = "Requests = " + str((request + 1) * 5)
    plt.plot(np.arange(6)[1:] * 5, makespan_mean_data[request, :], label = r_str, marker = "o")
    plt.fill_between(np.arange(6)[1:] * 5, makespan_mean_data[request, :] - makespan_std_data[request, :], makespan_mean_data[request, :] + makespan_std_data[request, :], alpha = 0.5)
    plt.xlabel("Number of blocks")
    plt.ylabel("Makespan")
ax = plt.gca()
for i, label in enumerate(ax.get_xticklabels()):
    if i % 2 == 0:
        label.set_visible(False)
plt.title("Makespan Benchmarks")
plt.legend()
```
